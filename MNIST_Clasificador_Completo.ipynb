## 2. Arquitectura Elegida <a id="2-arquitectura"></a>

### 2.1 Dise√±o de la Red Neuronal

Para este proyecto, he elegido una **Red Neuronal Multicapa (MLP)** con la siguiente arquitectura:

```
Entrada:      784 neuronas (28√ó28 p√≠xeles aplanados)
              ‚Üì
Capa Oculta 1: 128 neuronas + ReLU
              ‚Üì
Capa Oculta 2:  64 neuronas + ReLU
              ‚Üì
Salida:        10 neuronas + Softmax (probabilidades de cada clase)
```

### 2.2 Justificaci√≥n del Dise√±o

**¬øPor qu√© esta arquitectura?**

1. **Capa de entrada (784):** Cada p√≠xel de la imagen 28√ó28 es una caracter√≠stica
2. **Capa oculta 1 (128):** Suficiente capacidad para aprender patrones b√°sicos (bordes, curvas)
3. **Capa oculta 2 (64):** Refina y combina patrones de nivel inferior
4. **Reducci√≥n gradual:** 784‚Üí128‚Üí64‚Üí10 permite compresi√≥n progresiva de informaci√≥n
5. **ReLU:** Activaci√≥n no-lineal que permite aprender relaciones complejas
6. **Softmax:** Convierte scores en probabilidades interpretables

**Par√°metros totales:**
- W1: 784√ó128 = 100,352
- b1: 128
- W2: 128√ó64 = 8,192
- b2: 64
- W3: 64√ó10 = 640
- b3: 10
- **Total: 109,386 par√°metros**

### 2.3 Hiperpar√°metros

- **Learning rate:** 0.1
- **Batch size:** 128
- **√âpocas:** 30
- **Optimizador:** Gradient Descent (SGD)
- **Inicializaci√≥n:** He initialization para ReLU

### 6.1 Aprendizajes Principales
